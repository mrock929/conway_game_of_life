```python
import numpy as np

from game_of_life import GameOfLife
```

```python
# Blinker
board = np.zeros((6, 9))
board[1, 2] = 1
board[2, 2] = 1
board[3, 2] = 1
game = GameOfLife(board_array=board, show_plots=True, save_plots=False, file_prefix="blinker")
game.propagate_board(n_steps=4)
```

```python
# Block
board = np.zeros((4, 4))
board[1, 1] = 1
board[1, 2] = 1
board[2, 1] = 1
board[2, 2] = 1
game = GameOfLife(board_array=board, show_plots=True, save_plots=False, file_prefix="block")
game.propagate_board(n_steps=2)
```

```python
# Toad
board = np.zeros((6, 6))
board[2, 2] = 1
board[2, 3] = 1
board[2, 4] = 1
board[3, 1] = 1
board[3, 2] = 1
board[3, 3] = 1
game = GameOfLife(board_array=board, show_plots=True, save_plots=False, file_prefix="toad")
game.propagate_board(n_steps=4)
```

```python
# Glider
board = np.zeros((16, 18))
board[1, 2] = 1
board[2, 3] = 1
board[3, 1] = 1
board[3, 2] = 1
board[3, 3] = 1
game = GameOfLife(
    board_array=board, show_plots=True, save_plots=False, file_prefix="glider"
)
game.propagate_board(n_steps=20)
```
